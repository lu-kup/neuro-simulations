```python
import neuron_model

print(neuron_model.model)
print(neuron_model.neuron)

import numpy as np
from scipy.integrate import solve_ivp
import bifurcation_finder
import matplotlib.pyplot as plt
```

```python
ts = np.linspace(0, neuron_model.T, 10**5)
```

```python
neuron_model.I_app = 20
neuron_model.I_spike = 60

interval = 1
times = 4

spike_intervals = []
for i in range(times):
    current_time = 10 + i * interval
    spike_intervals.append([current_time, current_time + 0.5])

neuron_model.spike_intervals = spike_intervals
print(neuron_model.spike_intervals)

ax1, ax2, ax3 = neuron_model.visualise_phase_lines_triple(ts, (0,0,0.4), [[-68.64061, 0.01]], interval=400, annotation=True)

plt.show()
```

```python
def plot_spikes_sn(interval, times):
    spike_intervals = []
    for i in range(times):
        current_time = 10 + i * interval
        spike_intervals.append([current_time, current_time + 0.5])

    neuron_model.spike_intervals = spike_intervals
    print(neuron_model.spike_intervals)

    ax2, ax3 = neuron_model.visualise_phase_lines_double(ts, (0,0,0.4), [[-68.64061, 0.01]], interval=400, annotation=True)
    ax2.set_title(f"Intervalas = {interval}")
```

```python
for i in range(4):
    plot_spikes_sn(0.6 * 2**i, 6)
    plt.show()
```

```python
neuron_model.I_app = 35
neuron_model.I_spike = 100

frequency = 4
times = 7

spike_intervals = []
for i in range(times):
    current_time = 10 + i * frequency
    spike_intervals.append([current_time, current_time + 0.1])

neuron_model.spike_intervals = spike_intervals
print(neuron_model.spike_intervals)

V_sol = bifurcation_finder.steady_state_V(neuron_model.I_app, -80)
print(V_sol)
w_sol = neuron_model.w_inf(V_sol)
print(w_sol)

V_sol_unst = bifurcation_finder.steady_state_V(neuron_model.I_app, -40)
print(V_sol_unst)
w_sol_unst = neuron_model.w_inf(V_sol_unst)
print(w_sol_unst)

ax1, ax2, ax3 = neuron_model.visualise_phase_lines_triple(ts, (0,0,0.4), [[-64.0493, 0.01]], interval=400, annotation=True)
#ax1 = neuron_model.visualise_phase_lines(ts, 'black', [[-50, 0]], interval=400, annotation=True)
ax1.plot([V_sol], [w_sol], marker='o', markerfacecolor='black', markeredgecolor='black',zorder=10,markersize=10)
ax1.plot([V_sol_unst], [w_sol_unst], marker='o', markerfacecolor='white', markeredgecolor='black',zorder=10,markersize=10)

plt.show()
```
