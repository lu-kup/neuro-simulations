```python
# Saddle-node

config.model = "HT"

import neuron_model

print(neuron_model.model)
print(neuron_model.neuron)

import numpy as np
import matplotlib.pyplot as plt



def I_linear(t):
    return 30 + 0.25*t

neuron_model.I = I_linear
neuron_model.T = 100
ts = np.linspace(0, neuron_model.T, 10**5)


ax2, ax3 = neuron_model.visualise_phase_lines_double(ts, (0,0,0.4), [[-65, 0.0001]], interval=400, annotation=False, left_limit=-90)
plt.show()
```

```python
# Andronov-Hopf

config.model = "LT"

import neuron_model

print(neuron_model.model)
print(neuron_model.neuron)

import numpy as np
import matplotlib.pyplot as plt

def I_linear(t):
    return -30 + 0.25*t

neuron_model.I = I_linear
neuron_model.T = 100
ts = np.linspace(0, neuron_model.T, 10**5)


ax2, ax3 = neuron_model.visualise_phase_lines_double(ts, (0,0,0.4), [[-70, 0.0001]], interval=400, annotation=False, left_limit=-90)

plt.show()
```
