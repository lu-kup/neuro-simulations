```python
import neuron_model

print(neuron_model.model)
print(neuron_model.neuron)

import numpy as np
from scipy.integrate import solve_ivp
import bifurcation_finder
import matplotlib.pyplot as plt
```

```python
ts = np.linspace(0, neuron_model.T, 10**5)
```

```python
neuron_model.I_app = 0
neuron_model.I_spike = 0

V_sol = bifurcation_finder.steady_state_V(neuron_model.I_app, -80)
print(V_sol)
w_sol = neuron_model.w_inf(V_sol)
print(w_sol)

ax1, ax2 = neuron_model.visualise_phase_lines(ts, (0,0,0.4), [[-40, -0.1],[-50, 0.5],[-20, 0.7],[-30, 0.4], [-120,0.8], [-120,0.3], [-120,-0.15], [-130,0]], interval=1000, annotation=False, left_limit=-140)
#ax1 = neuron_model.visualise_phase_lines(ts, 'black', [[-50, 0]], interval=400, annotation=True)
ax1.plot([V_sol], [w_sol], marker='o', markerfacecolor='white', markeredgecolor='black',zorder=10,markersize=10)
ax2.remove()
ax1.set_title(f"I_app = {neuron_model.I_app}")
plt.show()
```
